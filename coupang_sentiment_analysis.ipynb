```python
def extract_nouns(text):
    text = text.strip()
    if not text:
        return []
    doc = nlp(text)
    nouns = []
    for sentence in doc.sentences:
        for word in sentence.words:
            lemma = word.lemma.split('+')
            xpos = word.xpos.split('+')
            
            for tok, pos in zip(lemma, xpos):
                if pos.startswith('n'):
                    nouns.append(tok)
    return nouns
```

```python
from hyperopt import hp
from hyperopt.pyll.stochastic import sample
```

```python
space = hp.choice('model', ['single', 'multi'])
```

```python
space = hp.choice('model', [
    {
        'layer': 'single',
        'lr': hp.loguniform('lr1', -7, 2) 
    },
    {
        'layer': 'multi', 
        'lr': hp.loguniform('lr2', -7, 2),
        'units': hp.quniform('units', 16, 128, 16), 
        'activation': hp.choice('activation', ['relu', 'tanh'])
    }    
])
```

```python
sample(space)
```

```python
import joblib
cv, x_train, x_test, y_train, y_test = joblib.load('review_coup.pkl')
```

```python
import tensorflow as tf
```

```python
from hyperopt import STATUS_OK
```

```python
import uuid
```

```python
str(uuid.uuid4())
```

```python
def objective(h):
    model = tf.keras.models.Sequential()
    
    if h['layer'] == 'multi':
        model.add(tf.keras.layers.Dense(16, activation=h['activation']))
    
    model.add(tf.keras.layers.Dense(1, activation='sigmoid'))
    
    model.compile(
        optimizer=tf.keras.optimizers.Adam(lr=h['lr']),
        loss='binary_crossentropy',
        metrics=['accuracy']
    )
    
    model.fit(
        x_train.toarray(),
        y_train.values,
        epochs=100,
        verbose=0
    )
    
    loss, acc = model.evaluate(x_test.toarray(), y_test.values, verbose=0)
    
    filename = str(uuid.uuid4())
    model.save(filename)
    return {
        'loss': -acc,
        'status': STATUS_OK,
        'filename': filename
    }
```

```python
def extract_nouns(text):
    text = text.strip()
    if not text:
        return []
    doc = nlp(text)
    nouns = []
    for sentence in doc.sentences:
        for word in sentence.words:
            lemma = word.lemma.split('+')
            xpos = word.xpos.split('+')
            
            for tok, pos in zip(lemma, xpos):
                if pos.startswith('n'):
                    nouns.append(tok)
    return nouns
```
